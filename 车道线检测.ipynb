```python
import cv2
import numpy as np
import os
```

```python

dispose_dir = 'img'  # 数据集文件夹
save_path = 'out'  # 输出文件夹
```

```python
for pic in os.listdir(dispose_dir):
    
    # 读取图片并二值化
    if pic.endswith('.jpg'):
        #print(os.path.join(dispose_dir, pic))
        img = cv2.imread(os.path.join(dispose_dir, pic))
        height, width, channel = img.shape
        
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#         gray2 = cv2.medianBlur(gray,15)
#         edges = cv2.Canny(gray2, 10, 100)
#         edges = cv2.dilate(edges,(3,3))
#         print(edges.shape)
#         gray=cv2.bitwise_or(gray,edges)
#         cv2.imwrite(os.path.join(save_path, "Hline"+pic), gray)  # 保存提交图片
        
        
        

        centerh = int(height/2)
        centerw = int(width/2)

        rgh=int(centerh/6)
        rgw=int(centerw/6)
        allmax=gray.mean()
        print("均值：{},长：{},宽:{},步长分别：{},{}".format(gray.mean(),height,width,rgh,rgw))
        
        
        for i in range(centerh,height,rgh):
            for j in range(0,width,rgw):
                x=pow(abs(i-centerh*1)/centerh,5)
                y=pow(abs(j-centerw*1)/centerw,2)
                distance = (1.0*x+2.0*y+1.0*x*y)+1.5
                now=gray[i:i+rgh,j:j+rgw]
                mavl = 0.5*allmax+0.8*now.mean()
                #mavl = 30+now.mean()+3*gray.mean()
                _,now_mask = cv2.threshold(now, mavl, 255, cv2.THRESH_BINARY)
                gray[i:i+rgh,j:j+rgw]=now_mask
                
        #cv2.imwrite(os.path.join(save_path, pic), gray)  # 保存提交图片
        
        gray=cv2.medianBlur(gray,5)
        
        
       
        kernel = np.ones((3,3),dtype = np.uint8)#定义卷积核
        #gray = cv2.erode(gray,kernel,iterations = 2)#
        
        
        
        gray = cv2.morphologyEx(gray, cv2.MORPH_OPEN, kernel)
        #gray = cv2.morphologyEx(gray, cv2.MORPH_CLOSE, k1)
        #gray=cv2.medianBlur(gray,3) 
        #gray = cv2.bitwise_or(gray,edges)
        
        cv2.imwrite(os.path.join(save_path, pic), gray)  # 保存提交图片
        
        

        
print("已完成")
```

```python
path="img"
car_mask = cv2.imread(dispose_dir+"/car_mask.png")
gray = cv2.cvtColor(car_mask, cv2.COLOR_BGR2GRAY)
_, car_mask = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY)
car_mask = (1-car_mask/255).astype('uint8')

for pic in os.listdir(dispose_dir):
    
    # 读取图片并二值化
    if pic.endswith('.jpg'):
        #print(os.path.join(dispose_dir, pic))
        img = cv2.imread(os.path.join(dispose_dir, pic))
        #print(type(img))
        height, width, channel = img.shape
        img[0:int(height/2), 0:width] = (0, 0, 0)        # 只分割图像的下面二分之一处
        img = cv2.bitwise_and(img, img, mask=car_mask)   # 删除汽车部分
        
        
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#         gray2 = cv2.medianBlur(gray,15)
#         edges = cv2.Canny(gray2, 10, 100)
#         edges = cv2.dilate(edges,(3,3))
#         print(edges.shape)
#         gray=cv2.bitwise_or(gray,edges)
#         cv2.imwrite(os.path.join(save_path, "Hline"+pic), gray)  # 保存提交图片
        
        
        

        centerh = int(height/2)
        centerw = int(width/2)

        rgh=int(centerh/6)
        rgw=int(centerw/6)
        allmax=0.0*gray.mean()
        print("均值：{},长：{},宽:{},步长分别：{},{}".format(gray.mean(),height,width,rgh,rgw))
        
        
        for i in range(centerh,height,rgh):
            for j in range(0,width,rgw):
                x=pow(abs(i-centerh*1)/centerh,5)
                y=pow(abs(j-centerw*1)/centerw,2)
                distance = (1.0*x+2.0*y+1.0*x*y)+1.5
                now=gray[i:i+rgh,j:j+rgw]
                mavl = allmax/distance+1.2*now.mean()
                #mavl = 30+now.mean()+3*gray.mean()
                _,now_mask = cv2.threshold(now, mavl, 255, cv2.THRESH_BINARY)
                gray[i:i+rgh,j:j+rgw]=now_mask
                
        #cv2.imwrite(os.path.join(save_path, pic), gray)  # 保存提交图片
        
        gray=cv2.medianBlur(gray,5)
        
        
        _,labels,stats,_ = cv2.connectedComponentsWithStats(gray, connectivity=8) 
        for st in range(len(stats)):
            if stats[st][4]>1000:
                labels[labels==st]=0
            else :
                labels[labels==st]=255
        gray=gray-labels
        gray[gray<0]=0
        gray=np.uint8(gray)
        
        
        
        
        #gray = cv2.morphologyEx(gray, cv2.MORPH_OPEN, (5,5))
        #gray = cv2.morphologyEx(gray, cv2.MORPH_CLOSE, k1)
        #gray=cv2.medianBlur(gray,3) 
        #gray = cv2.bitwise_or(gray,edges)
        
        cv2.imwrite(os.path.join(save_path, pic), gray)  # 保存提交图片
        
        

        
print("已完成")
```
